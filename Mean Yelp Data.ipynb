```python
import pandas as pd
import numpy as np

#Generate random integer values
import random
```

```python
csv_df1 = pd.read_csv("sf_restaurants.csv")
sf_restaurants_rating = np.asarray(csv_df1['rating'])
sf_restaurants_review = np.asarray(csv_df1['review_count'])

csv_df2 = pd.read_csv("sf_doctors.csv")
sf_doctors_rating = np.asarray(csv_df2['rating'])
sf_doctors_review = np.asarray(csv_df2['review_count'])

csv_df3 = pd.read_csv("sf_hotels.csv")
sf_hotels_rating = np.asarray(csv_df3['rating'])
sf_hotels_review = np.asarray(csv_df3['review_count'])

csv_df4 = pd.read_csv("phx_restaurants.csv")
phx_restaurants_rating = np.asarray(csv_df4['rating'])
phx_restaurants_review = np.asarray(csv_df4['review_count'])

csv_df5 = pd.read_csv("phx_hotels.csv")
phx_hotels_rating = np.asarray(csv_df5['rating'])
phx_hotels_review = np.asarray(csv_df5['review_count'])

csv_df6 = pd.read_csv("phx_doctors.csv")
phx_doctors_rating = np.asarray(csv_df6['rating'])
phx_doctors_review = np.asarray(csv_df6['review_count'])
```

```python
#SF Restaurants Rating and Review

rn_rating = []
rn_review = []
mean_rating = []
sf_rating_mean =[]
sf_review_mean =[]


for i in range(1000):
    rn_rating = random.sample(range(99),30)
    
    for j in rn_rating:
        mean_rating.append(sf_restaurants_rating[j])
    
    sf_rating_mean.append(np.mean(mean_rating))
    mean_rating = []


for i in range(1000):
    rn_review = random.sample(range(99),30)
    
    for j in rn_review:
        mean_rating.append(sf_restaurants_review[j])
    
    sf_review_mean.append(np.mean(mean_rating))
    mean_rating = []

#To CSV
df = pd.DataFrame({ 'Mean Rating': sf_rating_mean, 'Mean Review': sf_review_mean })
df.to_csv('sf_restaurants_mean.csv', index=False)
```
